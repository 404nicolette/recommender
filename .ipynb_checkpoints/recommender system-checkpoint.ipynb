```python
import pandas as pd
```

```python
# read the csv files
movies = pd.read_csv("movies.csv")
movies
```

```python
print(f"null values in the movie data frame: \n\n{movies.isnull().sum()}\n")
movies.info()
```

```python
ratings = pd.read_csv("ratings.csv")
ratings
```

```python
print(f"null values in the rating data frame: \n\n{ratings.isnull().sum()}\n")
ratings.info()

'''
both movies and ratings data frame have no null values.
'''
```

```python
merged = pd.merge(movies, ratings, on="movieId")
merged.head(10)

merged.isnull().sum()
```

```python
merged.drop(inplace=True, columns="timestamp")
merged
```

```python
# visualisation
import matplotlib.pyplot as plt

data = merged["rating"]

#plot
plt.hist(data, bins = 50, color="orange", edgecolor="black")
plt.xlabel("reviews")
plt.ylabel("movie count")
plt.title("ratings made on all movies")
plt.show()
```

```python
# filter
#content based filtering
split = merged["genres"].str.get_dummies(sep="|")
split
```

```python
main = pd.merge(merged, split, on=split.index)
main.drop(columns = ["key_0", "genres"], inplace=True)
main
```

```python
'''
binary scale simplifies the recommendation process by categorizing each rating into either positive (1) 
or negative (-1), making it easier to determine which items to recommend.
'''

def binary_rating(r):
    if r<4.0:
        return -1
    else:
        return 1
main.head(10)
```

```python
'''
create a dataframe with
columns set to all user IDs and the rows set to the movies’ ID
1 = positive ratings
-1 = negative ratings
0 = user did not rate the movie
'''

matrix = main.pivot(index="movieId", columns = "userId", values="rating")
matrix = matrix.fillna(0)
'''
NaN values represent movies that a user hasn't rated. By replacing NaN with 0, 
the system assumes that the absence of a rating means no preference, effectively 
treating it as a neutral or nonexistent rating.
'''
matrix
```

```python
# computing cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(matrix, matrix)

user = int(input("Enter movie id: "))
# get user to enter a movie title and if that movie title is present, find its associated 

# similarity to the first movie:
similar_to = pd.DataFrame(similarity[user, :], columns = ["similar"])
res = similar_to.sort_values(by="similar", ascending=False)
top_similar = res.head(11)
```

```python
temp = movies
temp.drop(columns = ["genres"], inplace=True)
```

```python
temp["title"] = temp["title"].str.replace(r'\s\(\d{4}\)', '', regex=True)
temp
```

```python
'''
get user to input movie title:
s1: remove the associated movie year
s2: make the title uniform[remove white space and make everything lowercase]
s3: match the title to the corresponding movieid
'''
top_similar_titles = temp.iloc[top_similar.index[1:11]]
print(f"The movies similar to {user} are: ")
print(top_similar_titles[["title"]])
```
